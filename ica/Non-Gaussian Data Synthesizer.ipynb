```python
np.random.seed(1)
```

```python
S = np.random.uniform(size=(4,200))
```

```python
for s in S:
    fig, ax = plt.subplots()
    ax.plot(s)
```

```python
A = np.random.exponential(size=(4,4))
```

```python
X = A@S
```

```python
for x in X:
    fig, ax = plt.subplots()
    ax.plot(x)
```

```python
ica = FastICA(n_components=4)
```

```python
S_ = ica.fit_transform(X.T)
```

```python
for s, s_ in zip(S, S_.T):
    fig, (ax1, ax2) = plt.subplots(ncols=2)
    ax1.plot(s_)
    ax2.plot(s)
```
